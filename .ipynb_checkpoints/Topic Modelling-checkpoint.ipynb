```python
from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import string
import pandas as pd
```

```python
df = pd.read_pickle('ted.pkl')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation,NMF
```

```python
tfidf = TfidfVectorizer(max_features=4000,stop_words='english')
```

```python
X = tfidf.fit_transform(df.Content)
```

```python
X.shape
```

```python
tfidf.get_feature_names()
```

```python
no_topics = 30

nmf = NMF(n_components=no_topics, random_state=1, alpha=.1, l1_ratio=.5, init='nndsvd')
X_nmf = nmf.fit(X)
```

```python
lda = LatentDirichletAllocation(n_topics=no_topics, max_iter=5, learning_method='online', \
                                learning_offset=50.,random_state=0)
X_lda = lda.fit(X)
```

```python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % (topic_idx))
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))

no_top_words = 10
display_topics(X_nmf, tfidf.get_feature_names(), no_top_words)
```

```python
display_topics(X_lda, tfidf.get_feature_names(), no_top_words)
```
